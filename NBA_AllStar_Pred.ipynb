```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import ExcelWriter

#Read csv file
#I have chosen columns that will be independent variables in the eventual ML model. They cannot be averages, as they will be incorrect values when later pivoting the data
cols = ['Year', 'Player', 'Pos', 'Tm', 'G', 'MP', 'PER', '3PAr', 'FTr','OWS','DWS','WS','FG','FGA','3P','3PA','2P','2PA','FT','FTA','ORB','DRB','TRB','AST','STL','BLK','PF','PTS']
stats = pd.read_csv ('/content/Seasons_Stats.csv', usecols = cols)

#The way this data is formatted, there is a line of data for every team of every season that a player has ever played and another total ("TOT") line. For example, if you played on the Celtics and Lakers in the same season, there will be 3 lines of data. To avoid double coutning stats, I will filter out all total lines.
stats = stats[stats['Tm'] != "TOT"]

#stats = stats[stats['Year'] < 2015]
#stats = stats[stats['Year'] >1980]


stats
```

```
#stats = pd.pivot_table(data=stats,index=['Player'],aggfunc={'G':np.sum,'MP':np.sum })

#Create aggregrate dataview
#Most columns are aggregrated by sum, others such as PER and winshares are by mean
stats = pd.pivot_table(data=stats,index=['Player'],aggfunc={'Year':np.mean, 'G':np.sum, 'MP':np.sum, 'PER':np.mean, '3PAr':np.sum, 'FTr':np.sum,'OWS':np.mean,'DWS':np.mean,'WS':np.mean,'FG':np.sum,'FGA':np.sum,'3P':np.sum,'3PA':np.sum,'2P':np.sum,'2PA':np.sum,'FT':np.sum,'FTA':np.sum,'ORB':np.sum,'DRB':np.sum,'TRB':np.sum,'AST':np.sum,'STL':np.sum,'BLK':np.sum,'PF':np.sum,'PTS':np.sum})
stats
```

```
#oldplayers = stats[stats['Year'] < 2009]
#oldplayers = oldplayers[oldplayers['Year'] >1980]
oldplayers
```

```
#Create new calculated columns

#Percentages
stats["FG%"] = stats["FG"] / stats["FGA"] 
stats["FT%"] = stats["FT"] / stats["FTA"] 
stats["3P%"] = stats["3P"] / stats["3PA"] 
stats["eFG%"] = (stats["FG"] + 5 * stats["3P"] ) / stats["FGA"] 

#Per game averages
stats["PPG"] = stats["PTS"] / stats["G"] 
stats["APG"] = stats["AST"] / stats["G"] 
stats["RPG"] = stats["TRB"] / stats["G"] 
stats["ORPG"] = stats["ORB"] / stats["G"] 
stats["DRPG"] = stats["DRB"] / stats["G"] 
stats["BPG"] = stats["BLK"] / stats["G"] 
stats["SPG"] = stats["STL"] / stats["G"] 
stats["FPG"] = stats["PF"] / stats["G"] 

#Per game averages
stats["PPM"] = stats["PTS"] / stats["MP"] 
stats["APM"] = stats["AST"] / stats["MP"] 
stats["RPM"] = stats["TRB"] / stats["MP"] 
stats["ORPM"] = stats["ORB"] / stats["MP"] 
stats["DRPM"] = stats["DRB"] / stats["MP"] 
stats["BPM"] = stats["BLK"] / stats["MP"] 
stats["SPM"] = stats["STL"] / stats["MP"] 
stats["FPM"] = stats["PF"] / stats["MP"] 

stats
```

```
#Select columns that will be used in ML model
stats = stats.loc[:, ('DWS',
 'OWS',
 'PER',
 'WS',
 'FG%',
 'FT%',
 '3P%',
 'eFG%',
 'PPG',
 'APG',
 'RPG',
 'BPG',
 'SPG',
 'ORPG',
 'DRPG',
 'FPG',
 'PPM',
 'APM',
 'RPM',
 'ORPM',
 'DRPM',
 'BPM',
 'SPM',
 'FPM')]
stats
```

```
HoF = pd.read_csv ('/content/HoF.csv')
HoF = HoF.rename(columns={'Inductees':'Player'})
HoF

#import list of all star players
allstar = pd.read_csv ('/content/allstar.csv')
allstar
```

```
df = pd.merge(stats,HoF,on='Player', how = 'inner', indicator = True)
df = pd.merge(stats,allstar,on='Player', how = 'left', indicator = True)
df
df['allstar'] = np.where(df['_merge']!= 'left_only', 1, 0)
df2 = df[df['allstar'] == 0]
df3 = df[df['allstar'] == 1]
#df3 = df2.sort_values('PER',ascending = False).groupby('Player').head(10)
#df2 = df2.set_index(['mainid','pidy']).groupby('pidx')['score'].nlargest(2).reset_index()
df2 = df2.nlargest(150, 'PER')

frames = [df2, df3]

result = pd.concat(frames)
result
```

```
#Select columns that will be used in ML model
dataset = result.loc[:, ('Player',
 'DWS',
 'OWS',
 'PER',
 'WS',
 'FG%',
 'FT%',
 '3P%',
 'eFG%',
 'PPG',
 'APG',
 'RPG',
 'BPG',
 'SPG',
 'ORPG',
 'DRPG',
 'FPG',
 'PPM',
 'APM',
 'RPM',
 'ORPM',
 'DRPM',
 'BPM',
 'SPM',
 'FPM',
 'allstar')]

dataset = dataset.replace([np.inf, -np.inf], np.nan)
dataset = dataset.fillna(0)
dataset
```

```
#DATA SCIENCE MACHINE LEARNING

X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```
#Encoding the dependent variable
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```
print(y)
```

```
#Split dataset into Training and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = .2, random_state = 1)
X_train
np.any(np.isnan(X_train))
np.isfinite(X_test.all()) 
```

```
#Feature scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
#Fit will get mean and SD, transform will apply mean and SD to the standardization formula


X_train [:,:] = sc.fit_transform(X_train [:,:])
X_test [:,:] = sc.transform(X_test [:,:])
X_train
```

```
#Logistic Regression
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
